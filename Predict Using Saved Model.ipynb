## Predict News Article Category Using Trained Data

#### Load Saved Model

```python
import pandas as pd
from sklearn.externals import joblib
from sklearn.feature_extraction.text import TfidfVectorizer

# Load The Model
loadedModel = joblib.load("TrainedData/Finalized Model.gz")

# Load The Category List
idToCategory = joblib.load("TrainedData/Category List.gz")

# Load Fitted TFIDF
TFIDFVectorizer = joblib.load("TrainedData/Fitted TFIDF.gz")
```

#### News Category Prediction

```python
# Add Texts in This Array
texts = [
    "Bangladesh Opener Tamim Iqbal's return to International cricket is likely to be delayed as he suffered a muscle strain on the right side of the rib.",
    "India may keep a ban on onion exports until February because domestic prices have risen after the harvest of summer-sown crops, which were expected to augment supplies, was delayed and damaged by untimely rains, a government official said.",
    "एडिलेड के मैदान पर ऑस्ट्रेलिया-पाकिस्तान के बीच डे-नाइट टेस्ट मैच खेला जा रहा है। इस मुकाबले में ऑस्ट्रेलिया के सलामी बल्लेबाज डेविड वार्नर ने तिहरा शतक जड़ा तो वहीं जब पाकिस्तान की टीम बल्लेबाजी करने उतरी तो मेजबान टीम के कप्तान टीम पेन ने कुछ ऐसा किया जिसे देखकर हर कोई हैरान रह गया। विकेटकीपर कप्तान ने कमाल का कैच लपका है। इसका वीडियो सोशल मीडिया पर काफी वायरल हो रहा है।",
    "बिग बॉस 13 के कंटेस्टेंट आसिम सोशल मीडिया पर हर दिन टॉप ट्रेंडिंग में नजर आ रहे हैं। लगातार ट्रेंड कर रहे आसिम अब BB13 में फैंस के फेवरेट कंटेस्टेंट बनकर उभर चुके हैं। आसिम को इतनी पॉपुलैरिटी मिली है जितनी सीजन 11 में शिल्पा शिंदे को मिली थी। ऐसे में आसिम #UnstoppableAsim के साथ शिल्पा शिंदे का भी रिकॉर्ड तोड़ चुके हैं। आसिम को लेकर कई सेलेब्स चर्चा करते देखे जा रहे हैं।",
    "বর্তমানে দেশের যেসব ব্যবসায়ী গ্রুপকে পেঁয়াজ আমদানি করতে বলা হয়েছে, তারা আগে এ ব্যবসায় জড়িত ছিল না। এতে করে আমদানিতে দেরি হচ্ছে এবং দাম কমতেও দেরি হচ্ছে বলে মনে করে কনজ্যুমার অ্যাসোসিয়েশন অব বাংলাদেশ (ক্যাব)।",
    "কুমিল্লার হয়ে ২০১৫ ও ২০১৮ বিপিএল চ্যাম্পিয়ন কোচ সালাউদ্দিন এবার নতুন চ্যালেঞ্জ নিয়ে শুরু করতে যাচ্ছেন টুর্নামেন্ট। ঢাকার দায়িত্ব এবার তাঁর হাতে। তাঁর প্রতি ঢাকার দর্শকদের প্রত্যাশা থাকবে অনেক বেশি। সালাউদ্দিন বলছেন, এ ধরনের চাপ সামলাতে অভ্যস্ত তিনি। তাঁর কাছে বরং ঢাকার চেয়ে কুমিল্লা ভিক্টোরিয়ানস সামলানো বেশি কঠিন মনে হয়েছে, ‘ঢাকার চেয়ে আমার কাছে মনে হয় কুমিল্লা দলে কোচিং করানো অনেক চাপের। কুমিল্লার সমর্থকদের উন্মাদনা বেশি থাকে। দলের প্রতি তাদের দরদও অনেক বেশি ছিল। এটা সামলানোর অভ্যাস আছে, এখানেও যদি চাপ থাকে আমরা সেটা সামলে নিতে পারব।",
]

# Predict and Print Results
for text, predicted in zip(texts, loadedModel.predict(TFIDFVectorizer.transform(texts))):
    print('"{}"'.format(text))
    print("  - Predicted Category: '{}'".format(idToCategory[predicted]))
    print("")
```

#### Bulk News Category Prediction

```python
# Load Data To Predict The News Category
inputDataFrame = pd.read_csv("DataSets/Random News Articles.csv")

# Predict from Model
features = TFIDFVectorizer.transform(inputDataFrame.Text.tolist())
yPrediction = loadedModel.predict(features)

# Since All Predictions are IDs, Need to Convert Back to Category Name
yPredictionName =[]
for categoryId in yPrediction:
    yPredictionName.append(idToCategory[categoryId])

#Create Full Output Dataframe
outputDataFrame = pd.DataFrame({
    "ActualCategory": inputDataFrame["Category"],
    "PredictedCategory": yPredictionName,
    "Title": inputDataFrame["Title"],
    "URL": inputDataFrame["URL"],
    "Text": inputDataFrame["Text"]
})
pd.set_option("display.max_rows", 125)
outputDataFrame.head(125)
```
